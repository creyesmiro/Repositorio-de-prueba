```python
## Obtención de base de datos
N=1000
mu, sigma = 0, 1 ## media y desviación estándar
X = np.random.normal(mu, sigma, N)
```

```python
## Histograma de distribución normal.
cuenta, cajas, ignorar = plt.hist(X, 20)
plt.ylabel('frequencia')
plt.xlabel('valores')
plt.title('Histograma')
plt.show()
```

```python
## Calcular valores de la función de distribución acumulada
def expression(t):
    return math.exp(-(t**2))

def Fp(X):
    res,err = quad(expression,0,X/np.sqrt(2))
    y = 1/2 * (1 + (2/np.sqrt(math.pi)) * res)
    return y
```

```python
## Graficar FUNCION DE DISTRIBUCIÓN ACUMULADA 
n = 50
X = np.linspace(0,n-1,n)
Y = []

for i in range(len(X)): 
    Y.append(Fp(X[i]))

plt.plot(X,Y)
```
